###  在项目中实践机器学习（Python版）

分类或回归模型的机器学习项目可以分为以下六个步骤：

1. 定义问题
    - 导入类库
    - 导入数据集
2. 理解数据
    - 描述性统计
    - 数据可视化
3. 数据准备
    - 数据清洗
        - 删除重复数据，标记错误数值
    - 特征选择
        - 去除多余的特征属性和增加新的特征属性
    - 数据转换
        - 对数据尺度进行调整，或者调整数据的分布
4. 评估算法
    - 分离数据集
        - 分理出评估数据集，以便于验证模型
    - 定义模型评估标准
        - 用来评估算法模型
    - 算法审查
        - 抽样审查线性算法和非线性算法
    - 算法比较
        -比较算法的准确度
5. 优化模型
    - 算法调参
        - 对每一种算法进行调参，得到最佳结果
    - 集成算法
        - 使用集成算法来提高算法模型的准确度
6. 结果部署
    - 预测评估数据集
    - 利用整个数据集生成模型
    - 序列化模型

### 使用模板的小技巧
- 快速执行一遍
    增加对项目的理解
- 循环
    多循环各个步骤，直到找到一个准确度足够的模型或达到预定的周期
- 尝试每一个步骤
    坚持在每个步骤中做些工作
- 定向准确度
    让每个步骤都要向着正向影响发展
- 按需适用
    按照项目需要修改步骤，把握每一次改进都以提高算法模型的准确度为前提的原则
